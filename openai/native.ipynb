```python
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

stream = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": "please tell me about a brief summary of air traffic control"}],
    stream=True,
)
for chunk in stream:
    print(chunk.choices[0].delta.content) 
     
```
